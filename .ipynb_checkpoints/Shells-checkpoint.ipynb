# Shells
This notebook will be concerned with taking the images of the camino shells, turning the image into a heightmap, and then placing hexagons over the top of it.

The hexagon coordinates can then be scaled, rotated, and placed underneath the heightmaps. This will allow for the relief of the camino de santiago shell to be placed in the 3d print.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import imageio as imagio

mpl.rc('image', cmap='terrain')


import loadData as LD
import CoordinateTransform as CT
import Hexagon as Hexagon
import HexGrid as HexGrid
```

### Loading in the image as a matrix

Firstly, we need to load in the image as a matrix to make sure that we can start placing the hexagons.

```python
im = imagio.imread('data\\shell\\shell.png')

print(im.shape)
```

```python
plt.imshow(im,origin='lower',interpolation='bilinear')
```

```python
fig = plt.figure()

for j in range(3):
    ax = fig.add_subplot(3,1,j+1)
    ax.imshow(im[:,:,j],origin='lower',interpolation='bilinear')

plt.show()
```

### Manipulating the image to a heightmap

I've found that the channels for RGB are 0,1,2 in the third dimension.
I've also demonstrated that channel [:,:,2] (blue) is the best indicator for where the shells are.

The image is also top-bottom reversed, which we need to undo.

```python
heightmap = np.flip(im[:,:,2],axis=0)

plt.imshow(heightmap,origin='lower',interpolation='bilinear')

print('min={} | max={}'.format(np.min(heightmap),np.max(heightmap)))
```

```python
heightmap = heightmap / 255
heightmap = np.around(heightmap)
plt.imshow(heightmap,origin='lower')
print('min={} | max={}'.format(np.min(heightmap),np.max(heightmap)))
print(np.unique(heightmap))
```

```python
# this roughly alligns the centre with where we want it to be over the image
scale = 540 * 2 / np.sqrt(3)
centre = np.array([520,508]).reshape(2,1)
rotation = -90

fig,ax = plt.subplots()
plt.hold = True

# plot underlying image and points on journey
ax.imshow(heightmap,origin='lower',interpolation='bilinear')

#--------------------------------------------------------------
hexcolour = 'r'

# create the initial hexagon, plot its vertices on the image
H = Hexagon.Hexagon(scale,rotation,centre)

plotVertices = np.hstack((H.vertices, H.vertices[:,0].reshape((2,1))))
ax.plot(plotVertices[0,:],plotVertices[1,:],lw=2,c=hexcolour)

'''
for j in range(5):
    H = H.createAdjacentHexagon(0)
    plotVertices = np.hstack((H.vertices, H.vertices[:,0].reshape((2,1))))
    ax.plot(plotVertices[0,:],plotVertices[1,:],lw=3,c=hexcolour)
'''
plt.show()
```

```python
# This will create a new heightmap that fully encloses the hexagon

shift = np.array([200,100]).reshape((2,))
sizeIncrease = np.array([500,500]).reshape((2,))

newHeightmap = np.zeros(tuple(np.array(heightmap.shape).reshape((2,))+sizeIncrease))

x = range(heightmap.shape[1])
y = range(heightmap.shape[0])
X,Y = np.meshgrid(x,y)

X += shift[0]
Y += shift[1]

newHeightMap[Y,X] = heightmap


#-----------------------------
#plotting
fig,ax = plt.subplots()
plt.hold = True

# plot underlying image and points on journey
ax.imshow(newHeightmap,origin='lower')

#--------------------------------------------------------------
hexcolour = 'r'

# create the initial hexagon, plot its vertices on the image
H = Hexagon.Hexagon(scale,rotation,centre + )

plotVertices = np.hstack((H.vertices, H.vertices[:,0].reshape((2,1))))
ax.plot(plotVertices[0,:],plotVertices[1,:],lw=2,c=hexcolour)

'''
for j in range(5):
    H = H.createAdjacentHexagon(0)
    plotVertices = np.hstack((H.vertices, H.vertices[:,0].reshape((2,1))))
    ax.plot(plotVertices[0,:],plotVertices[1,:],lw=3,c=hexcolour)
'''
plt.show()
```
